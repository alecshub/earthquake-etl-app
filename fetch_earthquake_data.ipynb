```python
import requests
import json
```

```python
url = 'https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.geojson'
response = requests.get(url)
print(response.status_code)

with open("earthquake_data.geojson", mode="wb") as file:
    file.write(response.content)
```

```python
# Preview the json data in pretty format
with open("earthquake_data.geojson", "r") as f:
    data = json.load(f)

for feature in data['features'][:3]:
    pretty_data = json.dumps(feature, indent=4)
    print(pretty_data)
```
